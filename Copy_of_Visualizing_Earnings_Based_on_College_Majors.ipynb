# Visualizing Earnings Based on College Majors

In this project, I will be analyzing a dataset containing information on college graduates between 2010 and 2012 and their respective job outcomes. The data was collected through a survey by American Community Survey and cleaned by FiveThirtyEight. I'm curious to break down the data and find out things like the earnings of popular majors and how popular majors are distributed by gender. My goal is to answer questions such as these while utilizing the pandas and Matplotlib libraries in order to effectively visualize the data and easily draw insights or conclusions.

```python
import pandas as pd
import matplotlib as plt
recent_grads = pd.read_csv('recent-grads.csv')
```

```python
%matplotlib inline
```

I'll start off by quickly exploring how the data looks in order to get familiar with the set.

```python
recent_grads.iloc[0]
```

The first thing I'm going to want to do is drop any rows with missing data, because missing values will cause matplotlib to throw errors. I can do this using the DataFrame.dropna() function, and I'll also compare the number of rows before and after I clean the missing data to see how much of the data set was affected.

```python
raw_data_count = len(recent_grads)
recent_grads.dropna(axis=0, inplace=True)
cleaned_data_count = len(recent_grads)
print(raw_data_count)
print(cleaned_data_count)
```

```python
size_to_median = recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Sample_size vs. Median')
size_to_urate = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample_size vs. Unemployment Rate')
fulltime_to_median = recent_grads.plot('Full_time', y='Median', kind='scatter', title='Full_time vs. Median')
women_to_urate = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='ShareWomen vs. Unemployment Rate')
men_to_median = recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs. Median')
women_to_median = recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median')
```

In the first two plots, it becomes increasingly clear that students in *less* popular majors tend to earn more money. This seems natural, as their skillsets will be harder to come by in their respective industries and hence the demand for these students will be higher. It's a safe conclusion to draw then that in order to maximize one's job prospects after graduation, one step that can be taken is choosing a less popular major. I also found a weak correlation between the percentage of women in a given major when compared to unemployment rate. That is, students that major in predominantly female majors do not tend to be more or less likely to find a job after graduation.
